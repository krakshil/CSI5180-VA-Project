```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_loss = [5.95589, 5.94227, 5.89658, 5.80723, 5.67242, 5.47324, 5.19043, 4.85822, 4.58687, 4.40579]
eval_loss = [5.96003, 5.93321, 5.86386, 5.75031, 5.58290, 5.33255, 5.00355, 4.68749, 4.38038, 4.32526]
epochs = list(range(1,11))
```

```python
plt.plot(epochs, train_loss, color="red", marker="o", label="Train Loss")
plt.plot(epochs, eval_loss, color="blue", marker="*", label="Eval Loss")
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss vs Epochs")
plt.show()
```
